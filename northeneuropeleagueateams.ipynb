```python
sns.set_style("whitegrid")
plt.figure(figsize=(15,15))
ulkeler = data_aer["COUNTRY"].unique()
colors = ["blue","red","green","yellow"]

for color, ulke in zip(colors,ulkeler):
        sns.displot(data_aer[data_aer["COUNTRY"] == ulke]["Average"])
        ulke_mean = data_aer[data_aer["COUNTRY"] == ulke]["Average"].mean()
        plt.vlines(ulke_mean, 0, 5.00, colors=color)
plt.legend(ulkeler)
```

## Scatterplot

```python
data_aer.head()
```

```python
nor_data = data_aer[(data_aer["COUNTRY"]=="Norway") & (data_aer["Founded"] > "1910")]
nor_data.head()
swe_data = data_aer[(data_aer["COUNTRY"]=="Sweden") & (data_aer["Founded"] > "1910")]
swe_data.head()
den_data = data_aer[(data_aer["COUNTRY"]=="Denmark") & (data_aer["Founded"] > "1910")]
den_data.head()
fin_data = data_aer[(data_aer["COUNTRY"]=="Finland") & (data_aer["Founded"] > "1910")]
fin_data.head()
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x="Founded",y="Average", data=nor_data)
sns.scatterplot(x="Founded",y="Average", data=swe_data)
sns.scatterplot(x="Founded",y="Average", data=den_data)
sns.scatterplot(x="Founded",y="Average", data=fin_data)
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x="Founded",y="Points", data=nor_data, hue= "Points")
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x="Founded",y="Points", data=nor_data, hue= "TEAM")
```

```python
nor_data.head()
```

```python
nor_data_means = nor_data.groupby(pd.to_datetime(nor_data['Founded'], errors='coerce', utc=True).dt.strftime('%Y-%m-%d')
).mean() 
nor_data_means = nor_data_means.reset_index()
nor_data_means.info()
```

```python
nor_data_means["Founded"] = nor_data_means["Founded"].astype("datetime64")
```

```python
nor_data_means.info()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x="Founded", y="Points", data=nor_data_means)
```

```python
data_1910 = data_aer[(data_aer["Founded"] > "1910")]
data_1910.head()
```

```python
data_mean = data_1910.groupby(["COUNTRY", pd.to_datetime(data_1910['Founded'], errors='coerce', utc=True).dt.strftime('%Y-%m-%d')
]).mean() 
data_means = data_mean.reset_index()
data_means["Founded"] = data_means["Founded"].astype("datetime64")
data_means.info()
```

```python
data_means
```

```python
plt.figure(figsize=(15,5))
for ulke in data_means["COUNTRY"].unique():
    data = data_means[data_means["COUNTRY"]== ulke]
    sns.lineplot(x="Founded", y="Points", data=data)
plt.legend(data_means["COUNTRY"].unique())
```
