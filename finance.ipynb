```python
d_2020 = qw[(qw["Date"] > "2020")]
d_2020.head()
```

```python
data_mean = d_2020.groupby(["Index", pd.to_datetime(d_2020['Date'], errors='coerce', utc=True).dt.strftime('%Y-%m-%d')
]).mean() 
data_means = data_mean.reset_index()
data_means["Date"] = data_means["Date"].astype("datetime64")
data_means.info()
```

```python
data_means
```

```python
plt.figure(figsize=(15,5))
for ex in data_means["Index"].unique():
    data = data_means[data_means["Index"]== ex]
    sns.lineplot(x="Date", y="Price", data=data)
plt.legend(data_means["Index"].unique())
```
